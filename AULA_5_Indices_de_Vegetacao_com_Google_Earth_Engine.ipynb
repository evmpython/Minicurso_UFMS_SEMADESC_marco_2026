Para detalhar melhor uma regi√£o geralmente plotamos o contorno daquela regi√£o no mapa. Assim o GEE tem uma base de dados dos contornos dos munic√≠pios para o mundo todo. Esta base de dados √© uma Feature Collection de [n√≠vel 2](https://developers.google.com/earth-engine/datasets/catalog/FAO_GAUL_2015_level2).

Primeiro selecionamos a base de dados global de munic√≠pios com a fun√ß√£o `ee.FeatureCollection('FAO/GAUL/2015/level2')` e depois escolhemos o munic√≠pio de **Campo Grande** aplicando a fun√ß√£o `filter(ee.Filter.eq('ADM2_NAME', 'Campo Grande'))`.

> **LEMBRE-SE**: Apenas a primeira letra de cada palavra dever ser ma√≠uscula e todas as letras devem estar sem acentos.  

```python
# shapefile de Campo Grande
campo_grande = ee.FeatureCollection('FAO/GAUL/2015/level2').filter(ee.Filter.eq('ADM2_NAME', 'Campo Grande'))
```

```python
# mostrando a FeatureCollection que cont√™m o contorno do munic√≠pio de Campo Grande
campo_grande
```

Iremos utilizar a fun√ß√£o `centerObject` para centralizar o nosso mapa em Campo Grande. Como argumento iremos inserir a *FeatureCollection* do munic√≠pio de Campo Grande e um zoom de valor 9. Al√©m disso, iremos utilizar a fun√ß√£o `addLayer` para plotar o contorno de Campo Grande no mapa.

```python
# criando um mapa interativo
Map = geemap.Map()

# centraliza o mapa na regi√£o
Map.centerObject(campo_grande, zoom=9)

# adicionando basemap
Map.add_basemap('SATELLITE')

# contorno do munic√≠pio de Campo Grande
style = {'color': 'yellow', 'fillColor': '00000000'}
Map.addLayer(campo_grande.style(**style), {}, 'Campo Grande')

# exibe na tela
Map
```

# **PARTE 2):** Conhecendo os √çndices Espectrais

![Texto alternativo](https://github.com/evmpython/Minicurso_UFCG_nov_2025/blob/main/logo/asi.png?raw=true)

Em sensoriamento remoto frequentemente combinamos algumas bandas espectrais dos sensores dos sat√©lites visando real√ßar as caracter√≠sticas de um determinado alvo que queremos estudar, por exemplo a vegeta√ß√£o. Ent√£o a combina√ß√£o de duas ou mais bandas dentro de uma equa√ß√£o, √© definido como `√çndices Espectrais`.

Existem diversos `√≠ndices espectrais` e com diversas aplica√ß√µes. Por exemplo, existem √≠ndices espectrais para indentificar: vegeta√ß√£o, corpos h√≠dricos, √°reas queimadas e etc. Pensando nisto o reposit√≥rio da [`Awesome Spectral Indices`](https://awesome-ee-spectral-indices.readthedocs.io/en/latest/index.html) disponibiliza a equa√ß√£o e as informa√ß√µes sobre o c√°lculo de [265 √≠ndices espectrais](https://github.com/awesome-spectral-indices/awesome-spectral-indices/blob/main/output/spectral-indices-table.csv) que podem ser empregados em estudos de sensoriamento remoto.





Por exemplo, o √≠ndice espectral mais amplamente empregado em sensoriamento remoto para estudar o estado da vegeta√ß√£o √© o `Normalized Difference Vegetation Index (NDVI)`. O NDVI emprega a banda do infravermelho pr√≥ximo (0,8 ¬µm, NIR) e vermelho (0,6 ¬µm, R) para real√ßar as caracter√≠sticas da vegeta√ß√£o numa determinada regi√£o. A seguinte equa√ß√£o √© utilizada para o c√°lculo do NDVI:


---


$$
NDVI = \frac{NIR ‚àí R}{NIR + R}
$$




---



Em que:
 - `NIR:` reflet√¢ncia no canal de 0,8 ¬µm
 - `R:` reflet√¢ncia no canal de 0,6 ¬µm  




*Os valores de NDVI tipicamente concentram-se entre -1 a 1, sendo que valores baixos indicam solo nu/pouca vegeta√ß√£o e valores mais altos indicam vegeta√ß√£o sadia. Veja a ilustra√ß√£o a seguir.*

---

1. Carregando a Cole√ß√£o de Imagens: `indices_vegetacao = ee.ImageCollection('MODIS/061/MOD13Q1')`
    - ee.ImageCollection(): Fun√ß√£o que acessa uma cole√ß√£o de imagens no GEE

    - 'MODIS/061/MOD13Q1': Caminho do dataset espec√≠fico:

        - MODIS: Sensor Terra/Aqua MODIS

    - Argumentos:

        - '2025-01-01': Data inicial (formato ISO: YYYY-MM-DD)

    - campo_grande: Objeto ee.Geometry (ponto, pol√≠gono ou ret√¢ngulo) representando a √°rea de interesse

4. Processamento das Imagens: `.map(lambda image: image.multiply(0.0001).copyProperties(image, ['system:time_start']))`
    - map(): Aplica uma fun√ß√£o a cada imagem da cole√ß√£o

        - image.multiply(0.0001): Multiplica os valores dos pixels por 0.0001 (os dados MODIS originais s√£o escalonados √ó10,000)

        - copyProperties(image, ['system:time_start']): Preserva os metadados originais, especificamente a propriedade de timestamp

## Antes e Depois: Split Map

Outro tipo de mapa interativo poss√≠vel de ser criado com `geemap` √© o chamado [**split map**](https://geemap.org/geemap/#geemap.geemap.Map.split_map), que nada mais √© que um painel que mostra duas imagens diferentes de uma √∫nica vez. Veja um exemplo [aqui](https://geemap.org/notebooks/04_split_panel_map/).


Vamos usar esse tipo de mapa para comparar o mapa de NDVI de `janeiro de 2024` e `janeiro de 2025` para o munic√≠pio de Campo Grande.



Carregando os dados

```python
# data de in√≠cio e fim
data_ini = '2014-01-01'
data_fim = '2026-01-01'

# carrega os dados
ndvi = ee.ImageCollection('MODIS/061/MOD13Q1') \
         .filterDate(data_ini, data_fim) \
         .filterBounds(campo_grande) \
         .select('NDVI') \
         .map(lambda image: image.multiply(0.0001).copyProperties(image, ['system:time_start']))
```

Extraindo a data das imagens

```python
# importa biblioteca pandas
import pandas as pd

# transforma para ano-mes-dia hora:minuto e extrae os valores
ndvi_2 = ndvi.map(lambda img: img.set( {"DATE": ee.Date(img.get("system:time_start")).format("YYYY-MM-dd hh:mm")}))

# supondo que voc√™ j√° tenha a lista de datas
dates_list = ndvi_2.aggregate_array("DATE").getInfo()

# criar o DataFrame
df_dates = pd.DataFrame({'DATE': dates_list})

# mostrar as primeiras linhas
df_dates
```

Agregando para a escala mensal. Estamos trabalhando com imagens de 2014 √† 2025, sendo que cada m√™s possui duas imagens (`Lembre-se`: a resolu√ß√£o temporal desse produto √© de 16 dias). Como iremos comparar as imagens dos meses, teremos que realizar um agrupamento das imagens para cada m√™s e depois calcular o valor m√°ximo de NDVI das imagens de um mesmo m√™s.

```python
# fun√ß√£o auxiliar para obter o primeiro dia de cada m√™s
def gerar_lista_meses(data_ini, data_fim):

    ini = ee.Date(data_ini)
    fim = ee.Date(data_fim)

    def gera_datas(n):
        return ini.advance(n, 'month')

    n_meses = fim.difference(ini, 'month').round()
    return ee.List.sequence(0, n_meses.subtract(1)).map(gera_datas)

# fun√ß√£o para calcular a m√©dia mensal
def media_mensal(data):

    # data atual
    data = ee.Date(data)

    # data do pr√≥ximo m√™s
    prox_mes = data.advance(1, 'month')

    # filtra o per√≠odo considerando o m√™s atual
    colecao_mes = ndvi.filterDate(data, prox_mes)

    # calcula a m√©dia para o per√≠odo
    media = colecao_mes.max().set({'system:time_start': data.millis(), 'mes': data.format('YYYY-MM')})

    return media
```

```python
# gera a data inicial e final de cada m√™s
datas_mensais = gerar_lista_meses(data_ini, data_fim)
datas_mensais
```

```python
# cria a ImageCollection com imagens mensais
ndvi_mensal = ee.ImageCollection(datas_mensais.map(media_mensal))
ndvi_mensal
```

Despois de realizado a agrega√ß√£o, as datas das imagens est√£o por m√™s.

```python
# importa biblioteca pandas
import pandas as pd

# transforma para ano-mes-dia hora:minuto e extrae os valores
ndvi_mensal_2 = ndvi_mensal.map(lambda img: img.set( {"DATE": ee.Date(img.get("system:time_start")).format("YYYY-MM-dd hh:mm")}))

# Supondo que voc√™ j√° tenha a lista de datas
dates_list = ndvi_mensal_2.aggregate_array("DATE").getInfo()

# Criar o DataFrame
df_dates = pd.DataFrame({'DATE': dates_list})

# Mostrar as primeiras linhas
df_dates
```

Agora, vamos conseguir criar o nosso novo mapa interativo com a fun√ß√£o `.split_map()` do `geemap`. Essa fun√ß√£o precisa de dois argumentos principais:
  - `left_layer()`: imagem que ficar√° na parte esquerda do mapa
  - `right_layer()`: imagem que ficar√° na parte direita do mapa

Al√©m disso, as duas camadas v√£o precisar invocar a fun√ß√£o `.geemap.ee_tile_layer()`. Vamos entender como funciona:

```python
# seleciona os dados de "Janeiro de 2024"
ndvi_jan24 = ndvi_mensal.filterDate('2024-01-01').median().clip(campo_grande)
```

```python
# seleciona os dados de "Janeiro de 2025"
ndvi_jan25 = ndvi_mensal.filterDate('2025-01-01').median().clip(campo_grande)
```

```python
# cria o objeto "Map"
Map = geemap.Map()

# centra o mapa na regi√£o
Map.centerObject(campo_grande, zoom=8)

# adicionar um basemap
Map.add_basemap('SATELLITE')

# criar o mapa interativo
Map.split_map(left_layer = geemap.ee_tile_layer(ndvi_jan24, param_vis, 'Janeiro de 2024'),
              right_layer = geemap.ee_tile_layer(ndvi_jan25, param_vis, 'Janeiro de 2025'))

# exibe na tela
Map
```

Agora que carregamos os dados no formato do GEE (isto √© *ImageCollection*) iremos converter esses dados para o formato de Dataset do xarray. Para realizar esta convers√£o utilizaremos a biblioteca Xee e xarray, e especificamente do xarray usaremos a fun√ß√£o `.open_dataset`, que serve para abrir um dataset multidimensional. Neste caso, indicaremos nesta fun√ß√£o alguns atributos, que s√£o:


---


`engine='ee'`:

Especifica que o Xarray deve usar o engine do Google Earth Engine para carregar os dados.

Isso requer que a biblioteca ee (Earth Engine) esteja instalada e autenticada.


---


`crs='EPSG:4326'`:

Define o sistema de refer√™ncia de coordenadas (CRS) para os dados.

EPSG:4326 √© o c√≥digo para WGS84 (coordenadas geogr√°ficas em latitude/longitude).

Isso garante que os dados ser√£o projetados corretamente.


---


`scale=0.00250`:

Define a resolu√ß√£o espacial em graus para os dados (aproximadamente 250 metros no equador). Lembre-se: **1 grau ~ 100 km ou 100.000 m**.

No GEE, a escala determina o tamanho do pixel na reproje√ß√£o.

Valores menores significam maior resolu√ß√£o.


---


`geometry=regiao`:

Especifica a regi√£o de interesse (ROI) para recortar os dados.

A roi deve ser um objeto geogr√°fico (como um pol√≠gono) com um m√©todo geometry().

Isso limita a extra√ß√£o de dados apenas √† √°rea especificada.


---



```python
# importando o xarray
import xarray as xr

# transformando para dataset do xarray
ds_ndvi = xr.open_dataset(ndvi,
                          engine = 'ee',
                          crs = 'EPSG:4326',
                          scale = 0.00250,
                          geometry = regiao)

# muda de "(time, lon, lat)" para "(time, lat, lon)"
ds_ndvi = ds_ndvi.transpose("time", "lat", "lon")

# aplica o offset
ds_ndvi = ds_ndvi * 0.0001
```

Carrega os dados

```python
# carrega os dados
ndvi = ee.ImageCollection('MODIS/061/MOD13Q1') \
         .filterDate('2015', '2026') \
         .filterBounds(regiao) \
         .select('NDVI') \
         .map(lambda image: image.multiply(0.0001).copyProperties(image, ['system:time_start']))

# quantidade de imagens da cole√ß√£o
ndvi.size()
```

Extrai a s√©rie temporal

```python
# fun√ß√£o para extrair a s√©rie temporal m√©dia
def extrair_serie_temporal(colecao, regiao, escala=250):

    def processar_imagem(imagem):

        data = ee.Date(imagem.get('system:time_start')).format('YYYY-MM-dd')

        ndvi_medio = imagem.reduceRegion(reducer = ee.Reducer.mean(),
                                         geometry = regiao,
                                         scale = escala,
                                         maxPixels = 1e9).get('NDVI')

        return ee.Feature(None, {'Data': data, 'NDVI': ndvi_medio})

    return colecao.map(processar_imagem)
```

```python
# extrair a s√©rie temporal
serie_temporal = extrair_serie_temporal(ndvi, regiao)
serie_temporal
```

```python
# importa a biblioteca pandas
import pandas as pd

# extrai as DATAS da FeatureCollection
datas = serie_temporal.aggregate_array('Data').getInfo()

# extrai os DADOS de ndvi da FeatureCollection
valores = serie_temporal.aggregate_array('NDVI').getInfo()

# converte para DataFrame
df = pd.DataFrame({'Data': pd.to_datetime(datas), 'NDVI': valores}).sort_values('Data').reset_index(drop=True)

# remove valores nulos (caso existam)
df = df.dropna()

# mostra os dados
df
```

## Carregando os dados, aplicando o mascaramento de nuvens e calculando o NDVI

Iremos carregar os dados de `junho, julho e agosto de 2025` do sat√©lite [Sentinel-2](https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S2_SR_HARMONIZED?hl=pt-br) para calcular o NDVI. Al√©m disso iremos carregar a base de dados de [probabiblidade de nuvens](https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S2_CLOUD_PROBABILITY) para fazermos o mascaramento de nuvens. Lembrando que as imagens Sentinel-2 possuem frequ√™ncia temporal de 5 dias.



```python
# carregando os dados
sen2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED").linkCollection(ee.ImageCollection("COPERNICUS/S2_CLOUD_PROBABILITY"), 'probability') \
         .filterDate('2025-06-01', '2025-09-01') \
         .filterBounds(regiao)
sen2
```

A fun√ß√£o a seguir aplica o mascaramento de nuvens e depois calcula o NDVI. Das imagens de Junho, Julho e Agosto de 2025 utilizaremos apenas as imagens que possuem menos de 20% de cobertura de nuvens, depois aplicamos o fator de escala (0,0001) e calculamos o NDVI.

No sat√©lite Sentinel-2 as bandas do infravermelho pr√≥ximo e vermelho s√£o B8 (835,1 nm - NIR) e B4 (664,5 nm - Vermelho), respectivamente.

Ent√£o o c√°lculo do NDVI para as bandas do sat√©lite Sentinel-2 fica:

            NDVI = (B8 - B4) / (B8 + B4)

 - B8 = 835,1 nm - NIR
 - B4 = 664,5 nm - Vermelho

```python
# criando uma fun√ß√£o que aplica o mascaramento de nuvens e calcula o NDVI
def cloud_mask(img):

    # seleciona a banda que tem a informa√ß√£o sobre a porcentagem de nuvens das imagens
    qa = img.select('probability')

    # aplica mascara de nuvens. S√£o utilizadas apenas as imagens que possuem menos de 20% de cobertura de nuvens
    cloud_mask = qa.lt(20)

    # seleciona todas as bandas come√ßando com "B" e aplica fator de escala
    ms = img.select('B.*').multiply(0.0001)

    # calculando NDVI
    ndvi = ms.normalizedDifference(['B8','B4']).rename('ndvi')

    return ndvi.updateMask(cloud_mask).copyProperties(img, ['system:time_start'])
```

```python
# carregando os dados com a aplica√ß√£o da m√°scara de nuvens
sen2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED").linkCollection(ee.ImageCollection("COPERNICUS/S2_CLOUD_PROBABILITY"), 'probability') \
         .filterDate('2025-06-01', '2025-09-01') \
         .filterBounds(regiao) \
         .map(cloud_mask)
sen2
```

# **PARTE 9):** NDVI com os Sat√©lites da S√©rie `Landsat`

![Texto alternativo](https://github.com/evmpython/Minicurso_UFMS_SEMADESC_marco_2026/blob/main/04_logos/5019-Landsat-timeline-L7_decomissioned_2025_1.png?raw=true)

Desde 1972, os sat√©lites [Landsat](https://www.usgs.gov/landsat-missions/landsat-satellite-missions) t√™m adquirido continuamente imagens espaciais da superf√≠cie terrestre, fornecendo dados ininterruptos para auxiliar a gest√£o ambiental e decisores pol√≠ticos a tomar decis√µes sobre os recursos naturais e o meio ambiente. O primeiro sat√©lite da s√©rie foi o Landsat-1 lan√ßado em 1972 e o √∫ltimo √© o Landsat-9 em 2021. A grande vantagem da s√©rie Landsat √© a grande s√©rie hist√≥rica de dados e a alta resolu√ß√£o espacial (30 m), possuindo uma resolu√ß√£o temporal m√©dia de 16 dias. Maiores detalhes sobre os sensores dos sat√©lites Landsat voc√™ pode encontrar no site do [Observing Systems Capability Analysis and Review Tool](https://space.oscar.wmo.int/satelliteprogrammes/view/landsat) da WMO.


Recentemente o Google Earth Engine (GEE) disponibilizou v√°rias [cole√ß√µes](https://developers.google.com/earth-engine/datasets/catalog/landsat?hl=pt-br) de composi√ß√µes combinando todos os sat√©lites da s√©rie Landsat. Dentre estas cole√ß√µes temos os √≠ndices espectrais BAI, EVI, NBR, NDVI e NDWI em cad√™ncias de 8 dias, 32 dias e anuais.

Vamos utilizar os [dados](https://developers.google.com/earth-engine/datasets/catalog/LANDSAT_COMPOSITES_C02_T1_L2_32DAY_NDVI?hl=pt-br) de NDVI com frequ√™ncia temporal de 32 dias para analisar a s√©rie temporal de NDVI para uma localidade do Estado do MS. Iremos utilizar como refer√™ncia a latitude de -20.89 e longitude de -54.27 e aplicar com um raio de abrang√™ncia (buffer) de 100 m.



```python
#=============================================================================================#
#                                  IMPORTA BIBLIOTECAS
#=============================================================================================#
import pandas as pd
import matplotlib.pyplot as plt

#=============================================================================================#
#                                  CARREGA OS DADOS
#=============================================================================================#
# define a regio√£o: buffer de 10.000 metros
regiao = ee.Geometry.Point([-54.27, -20.89]).buffer(100) # (longitude, latitude) com buffer de 10.000 metros

# carrega os dados
ndvi = ee.ImageCollection('LANDSAT/COMPOSITES/C02/T1_L2_32DAY_NDVI') \
         .filterDate('1984', '2027') \
         .filterBounds(regiao) \
         .select('NDVI')

#=============================================================================================#
#                                 EXTRAI S√âRIE TEMPORAL
#=============================================================================================#
# fun√ß√£o para extrair a s√©rie temporal m√©dia
def extrair_serie_temporal(colecao, regiao, escala=30):

    def processar_imagem(imagem):

        data = ee.Date(imagem.get('system:time_start')).format('YYYY-MM-dd')

        ndvi_medio = imagem.reduceRegion(reducer = ee.Reducer.mean(),
                                         geometry = regiao,
                                         scale = escala,
                                         maxPixels = 1e9).get('NDVI')

        return ee.Feature(None, {'Data': data, 'NDVI': ndvi_medio})

    return colecao.map(processar_imagem)

# extrai a s√©rie temporal
serie_temporal = extrair_serie_temporal(ndvi, regiao)

# elimina os valores de datas/NDVI onde o NDVI for NaN
serie_temporal = serie_temporal.filter(ee.Filter.notNull(['NDVI']))

# extrai as DATAS da FeatureCollection
datas = serie_temporal.aggregate_array('Data').getInfo()

# extrai os DADOS de ndvi da FeatureCollection
valores = serie_temporal.aggregate_array('NDVI').getInfo()

# converte para DataFrame
df = pd.DataFrame({'Data': pd.to_datetime(datas), 'NDVI': valores}).sort_values('Data').reset_index(drop=True)

# remove valores nulos (caso existam)
df = df.dropna()

#=============================================================================================#
#                                  PLOTA FIGURA
#=============================================================================================#
# cria a moldura da figura
plt.figure(figsize=(14, 7))

# plota curvas
plt.plot(df['Data'], df['NDVI'], 'g-', linewidth=1, alpha=0.6)
plt.scatter(df['Data'], df['NDVI'], c=df['NDVI'], cmap='YlGn', s=20)

# configura√ß√£o dos eixos do gr√°fico
plt.title('S√©rie Temporal de NDVI - LANDSAT: 1984-2026')
plt.ylabel('NDVI')
plt.xlabel('Data')
plt.colorbar(label='NDVI')
plt.grid(True, linestyle='--', alpha=0.7)

# salva a figura
plt.savefig('Parte_9_serie_temporal_ndvi_LANDSAT.png', dpi=300, bbox_inches='tight')

# mostra a figura
plt.show()
```

Bom, ent√£o como podemos calcular os √≠ndices espectrais com o spyndex?

 - Para calcular os √≠ndices usaremos o m√©todo `computeIndex()`. Com ele podemos calcular quantos √≠ndices espectrais desejarmos!

 - O par√¢metro `index` recebe o √≠ndice espectral ou uma lista de √≠ndices espectrais a serem calculados, enquanto o par√¢metro `params` recebe um dicion√°rio com os par√¢metros necess√°rios para o c√°lculo dos √≠ndices espectrais.

Veja o exemplo a seguir:

Carrega o shapefile do munic√≠pio de Campo Grande

```python
campo_grande = ee.FeatureCollection('FAO/GAUL/2015/level2').filter(ee.Filter.eq('ADM2_NAME', 'Campo Grande'))
```

Carrega os dados

```python
sen2 = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
         .filterDate('2025-01-01', '2026-01-01') \
         .filterBounds(campo_grande) \
         .select('B.*') \
         .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE',10)) \
         .map(lambda image: image.multiply(0.0001).copyProperties(image, ['system:time_start']))
```

Calcula os √≠ndices

```python
# calcula a mediana e extrai os dados para √°rea de interesse
sen2 = sen2.median().clip(campo_grande)

# calcula os √≠ndices
idx = spyndex.computeIndex(index = ["VARI", "NDWI", "NDVI"],
                           params = {"R": sen2.select("B4"),
                                     "G": sen2.select("B3"),
                                     "B": sen2.select("B2"),
                                     "N": sen2.select("B8")})

# adiciona ao dataset os √≠ndices espectrais calculados
img = sen2.addBands(idx)
img
```

Plota o mapa interativo com os dados

```python
# cria a moldura do mapa
Map = geemap.Map()

# centraliza o mapa na regi√£o
Map.centerObject(campo_grande, zoom=9)

# contorno de Campina Grande
Map.addLayer(campo_grande, {}, 'Campo Grande')

# mapas
Map.addLayer(img['NDVI'].clip(campo_grande), {'min':-1, 'max':1, 'palette': ['blue', 'white', 'green']}, 'NDVI')
Map.addLayer(img['NDWI'].clip(campo_grande), {'min':-1, 'max':1, 'palette': ['red', 'white', 'green']}, 'NDWI')
Map.addLayer(img['VARI'].clip(campo_grande), {'min':-1, 'max':1, 'palette': ['red', 'white', 'green']}, 'VARI')

# barra de cores
Map.add_colorbar({'min':-1, 'max':1, 'palette': ['blue', 'white', 'green']}, label='NDVI', orientation='vertical')
Map.add_colorbar({'min':-1, 'max':1, 'palette': ['red', 'white', 'green']}, label='NDWI', orientation='vertical')
Map.add_colorbar({'min':-1, 'max':1, 'palette': ['red', 'white', 'green']}, label='VARI', orientation='horizontal')

# exibe na tela
Map
```
